# 전체 단어 워드클라우드

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import platform
```

```python
Seoul=pd.read_excel('서울 200806 통합.xlsx',index_row=0)
```

```python
def wordcloudf(dataframe):
    wordcloud = WordCloud(font_path='C:/Users/gram/공포논문/malgun.ttf',
                      background_color = 'black', #배경색 
                      width = 800, height = 600).generate(' '.join(dataframe['파일이름'])) 

    plt.figure(figsize = (15, 10)) # (가로인치, 세로인치) 
    plt.axis("off") # 축눈금 제거 
    plt.imshow(wordcloud) # 이미지가 표시되도록 
    return plt.show() # 최종 출력문
```

```python
wordcloudf(Seoul)
```
